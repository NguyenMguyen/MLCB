# activation function

```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
```

```python
s = 7
x = np.linspace(-s, s, 100)

z1 = np.exp(x)
z2 = np.exp(-x)
sigmoid = 1/(1 + z2)
tanh = (z1 - z2)/(z1 + z2)

nrows = 1
ncols = 2
width = 6*ncols
height = 3*nrows
plt.close('all')
fig, axs = plt.subplots(nrows, ncols, figsize=(width, height))

axs[0].plot(x, sigmoid)
axs[0].plot([-s, s], [0, 0], '-', color='orange')
axs[0].plot([-s, s], [1, 1], '-', color='orange')
# axs[0].set_title('$y = \frac{1}{1 + exp(-z)}$')
axs[0].set_xlabel('$z$', fontsize=18)
axs[0].set_ylabel('$a = 1/(1 + e^{-z})$', fontsize=18)
axs[0].tick_params(axis='both', which='major', labelsize=18)

axs[1].plot(x, tanh)
axs[1].plot([-s, s], [0, 0], '-', color='k')
axs[1].plot([-s, s], [-1, -1], '-', color='orange')
axs[1].plot([-s, s], [1, 1], '-', color='orange')
axs[1].tick_params(axis='both', which='major', labelsize=18)
axs[1].set_xlabel('$z$', fontsize=14)
axs[1].set_ylabel('$a = (e^{z} - e^{-z}/(e^{z} + e^{-z})$', fontsize=18)
# filename = 'sigmoid_tanh.pdf'
# with PdfPages(filename) as pdf:
#     pdf.savefig(bbox_inches='tight')
plt.show()
```

```python
plt.plot(x, sigmoid, color = 'blue')
plt.plot([-s, s], [0, 0], '-', color = 'orange')
plt.plot([-s, s], [1, 1], '-', color = 'orange')
# axs[0].set_title('$y = \frac{1}{1 + exp(-z)}$')
plt.xlabel('$z$', fontsize = 18)
plt.ylabel('$a = 1/(1 + e^{-z})$', fontsize = 18)
plt.title('sigmoid function', fontsize = 16)
plt.tick_params(axis='both', which='major', labelsize=18)

# filename = 'sigmoid.pdf'
# with PdfPages(filename) as pdf:
#     pdf.savefig(bbox_inches='tight')
plt.show()
```

```python
plt.plot(x, tanh, color = 'blue')
plt.plot([-s, s], [0, 0], '-', color = 'k')
plt.plot([-s, s], [-1, -1], '-', color = 'orange')
plt.plot([-s, s], [1, 1], '-', color = 'orange')
# axs[0].set_title('$y = \frac{1}{1 + exp(-z)}$')
plt.xlabel('$z$', fontsize = 18)
plt.ylabel('$a = (e^{z} - e^{-z}/(e^{z} + e^{-z})$', fontsize = 18)
plt.title('tanh function', fontsize = 16)
plt.tick_params(axis='both', which='major', labelsize=18)

# filename = 'tanh.pdf'
# with PdfPages(filename) as pdf:
#     pdf.savefig(bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(-3, 3, 100)
r = np.maximum(0, x)
s = 3
plt.plot([-s, 3], [0, 0], '-', color = 'k')
plt.plot(x, r, color = 'b')
# plt.axis('equal')
plt.xlabel('$z$', fontsize = 16)
plt.ylabel('$a = \max(z, 0)$', fontsize = 18)
plt.title('ReLU function', fontsize = 18)
# filename = 'relu.pdf'
# with PdfPages(filename) as pdf:
#     pdf.savefig(bbox_inches='tight')
plt.show()
```

```python
plt.subplot(121)

plt.plot([0, 0, 1], [0, 1, 0], 'ro', markersize=10, alpha=1)
plt.plot([1], [1], 'bs', markersize=10, alpha=1)

# plt.text(3.6, 2.5, r'?', fontsize=15)
# hide tikcs
cur_axes = plt.gca()
cur_axes.axes.get_xaxis().set_ticks([0, 1])
cur_axes.axes.get_yaxis().set_ticks([0, 1])

plt.xlabel('$x_1$', fontsize=20)
plt.ylabel('$x_2$', fontsize=20)

# fill region colors
x1 = np.arange(-10, 10, 0.1)
y1 = 1.5 - x1
y2 = 10
y3 = -10
plt.plot(x1, y1, 'k')
plt.fill_between(x1, y3, y1, color='red', alpha=0.1)
plt.fill_between(x1, y1, y2, color='blue', alpha=.1)

plt.axis('equal')
plt.ylim(-.5, 1.5)
plt.xlim(-.5, 1.5)
plt.title('AND problem')
plt.text(0, 0.5, r'$2x_1 + 2x_2 - 3 = 0$', fontsize=14, rotation=-45)
plt.text(-.3, 1.78, r'+', fontsize=15)
plt.text(-.3, 1.55, r'-', fontsize=20)
# plt.savefig('AND.png', bbox_inches='tight', dpi = 500)
# filename = 'AND.pdf'
# with PdfPages(filename) as pdf:
#     pdf.savefig(bbox_inches='tight')
plt.show()
```
