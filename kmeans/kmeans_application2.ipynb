# segmentation and compression

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = mpimg.imread('girl3.jpg')
plt.imshow(img)
plt.axis('off')
plt.show() 
```

```python
X = img.reshape(img.shape[0]*img.shape[1], img.shape[2])
```

## số lượng cluster khác nhau

```python
for K in [2, 3, 5, 10, 15, 20, 30]:
    kmeans = KMeans(n_clusters=K).fit(X)
    label = kmeans.predict(X)

    img1 = np.zeros_like(X)

    # replace each pixel by its center
    for k in range(K):
        img1[label==k] = kmeans.cluster_centers_[k]
    
    # reshape and display output image
    print('K = ', K)
    fn = 'girl_K' + str(K) 
    img2 = img1.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(img2, interpolation='nearest')
    plt.axis('off')
    plt.savefig(fn, bbox_inches='tight', dpi=400)
    plt.show()
```
